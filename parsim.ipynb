```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://coronavirus-graph.ru/mir'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

table = soup.find('table')
rows = table.find_all('tr')
result = []

for row in rows:
    cols = row.find_all('td')
    cols = [col.text.strip() for col in cols]
    result.append(dict(zip(['Дата', 'Заражений всего', 'Умерло всего', 'Летальность', 'Выздоровело всего', 'Болеющих'], cols)))
```

```python
# создаем списки где будут храниться даты и число болеющих конкретный день
date = []
boleyt = []

for i in result:
    #print(i, sep = " ", end = '\n')
    for j in i:
        if j == 'Дата':
            date.append(i[j])
        if j == 'Болеющих':
            boleyt.append(i[j])
        #print(j, i[j], sep = " ", end = '\n')
```

```python
# напишем функцию, которая преобразует данные вида datestring в вид гггг-мм-дд
datestring = "31 мая 2023"

d = {'января': '01',
     'февраля': '02',
     'марта': '03' ,
     'апреля': '04', 
     'мая': '05',
     'июня': '06',
    'июля': '07',
     'августа': '08',
     'сентября': '09',
     'октября': '10',
     'ноября': '11',
     'декабря': '12' }

def dat(ds):
    l = list(ds.split(' '))
    otvet = l[2] + '-' + d[l[1]] + '-' + l[0]
    return otvet

print(dat(datestring))
```

```python
# создаем новый список с изменённным способом записи даты
data = []
i = 2023
for j in date:
    if j == '31 декабря':
        i -= 1
    j += ' '
    j += str(i)
    data.append(dat(j))
```

```python
# данные в списке boleyt сохранились диковинным образом. '-' и '+' показывают динамику относительно предыдущего дня.
# Из них нужно вычленить число до знака,  этим и занимается код в нижнем окне. 
boleyt
```

```python
boleyt_int = []
for i in boleyt:
    i = i[1:]
    i = i.replace('-', '!').replace('+', '!').replace(' ', '')
    k = ''
    flag = True
    for z in i:
        if z == '!':
            flag = False
        elif z != '!' and flag:
            k += z
    i = int(k)
    boleyt_int.append(i)

for i in range(0, len(data)):
    print(data[i], boleyt_int[i], sep = " ", end = '\n')
```

```python
import pandas as pd

df = pd.DataFrame({'Дата': data, 'Количество заболевших COVID-19, чел.': boleyt_int})

display(df)
```

```python
# сохраняем наш получившийся датасет в файл covid.csv
df.to_csv('covid.csv', index=False)
```
